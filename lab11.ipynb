```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
df.isnull().sum()
```

```python

df_drop = df.dropna(subset=['age'])
print(df_drop.shape)
```

```python
df.fillna({'age': df['age'].mean()}, inplace=True)
print(df)
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

print(df)
```

```python
duplicates = df.duplicated().sum()
print("Duplicate rows:", duplicates)
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower()
print(df.columns)
```

```python
df['name'] = df['who'].astype(str) + "_@#"
df['name'] = df['name'].str.replace('[^A-Za-z0-9 ]+', '', regex=True)
df['name'].head()
```

```python
df.loc[:, 'embarked'] = df['embarked'].astype(str).str.strip()
print(df['embarked'].head())
```

```python
df.loc[:, 'sex'] = df['sex'].str.capitalize()
print(df['sex'].head())
```

```python
df.loc[:, 'pclass'] = df['pclass'].astype('category')
print(df['pclass'].head())
```

```python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['age'] >= Q1 - 1.5*IQR) & (df['age'] <= Q3 + 1.5*IQR)]
print(df['age'].describe())

```

```python
invalid_fare = df[df['fare'] < 0]
print(invalid_fare)
```

```python
invalid_age = df[(df['age'] < 0) | (df['age'] > 100)]
print(invalid_age)
```

```python
def validate_dataset(df):
    print("Nulls:", df.isnull().sum().sum())
    print("Duplicates:", df.duplicated().sum())

    if 'age' in df.columns:
        print("Age range:", df['age'].min(), "-", df['age'].max())
    elif 'Age' in df.columns:
        print("Age range:", df['Age'].min(), "-", df['Age'].max())

validate_dataset(df)
```

```python
def clean_data(data):
    data = data.copy()

    data.columns = data.columns.str.strip().str.lower()

    data.drop_duplicates(inplace=True)

    data['age'] = data['age'].fillna(data['age'].mean())

    data['embarked'] = data['embarked'].fillna(
        data['embarked'].mode()[0]
    )

    data['sex'] = data['sex'].astype(str).str.capitalize()

    return data


df_clean = clean_data(df)

print(df_clean.head())
```

```python
df_clean.to_csv('cleaned_titanic.csv', index=False)

print("Cleaned dataset saved successfully!")
```

```python
import numpy as np

# Detect Outliers using IQR
Q1 = df_clean['fare'].quantile(0.25)
Q3 = df_clean['fare'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Mean of fare column
fare_mean = df_clean['fare'].mean()

# Replace outliers with mean
df_clean['fare'] = np.where(
    (df_clean['fare'] < lower_bound) |
    (df_clean['fare'] > upper_bound),
    fare_mean,
    df_clean['fare']
)

print("Outliers replaced successfully!")
```

```python
if df_clean.isnull().sum().sum() == 0 and df_clean.duplicated().sum() == 0:
    print("Data is clean!")
else:
    print("Dataset still contains nulls or duplicates.")
```
