```python
import kagglehub
import pandas as pd
import os

# Download dataset
path = kagglehub.dataset_download("yasserh/titanic-dataset")

# Load CSV
df = pd.read_csv(os.path.join(path, "Titanic-Dataset.csv"))

# Check data
print(df.head())
print("\nShape:", df.shape)
print("\nData Types:")
print(df.dtypes)
```

```python
import matplotlib.pyplot as plt


# 2. Statistical Measures for Numeric Columns
numeric_cols = df.select_dtypes(include=['int64', 'float64'])

stats = pd.DataFrame({
    'Mean': numeric_cols.mean(),
    'Median': numeric_cols.median(),
    'Std Dev': numeric_cols.std(),
    'Min': numeric_cols.min(),
    'Max': numeric_cols.max(),
    'Skewness': numeric_cols.skew(),
    'Kurtosis': numeric_cols.kurt()
})

print("\n===== Statistical Summary =====")
print(stats)


# 3(a). Histogram of Passenger Ages
plt.figure(figsize=(8,5))
plt.hist(df['Age'].dropna(), bins=20)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()


# 3(b). Bar Chart of Survival Counts by Passenger Class
survival_class = pd.crosstab(df['Pclass'], df['Survived'])

survival_class.plot(kind='bar', figsize=(8,5))
plt.title('Survival Counts by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
